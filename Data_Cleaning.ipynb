```python
import pandas as pd
import numpy as np
```

```python
#load messy data file
df = pd.read_csv("marketing_campaign_data_messy.csv")
print(f"Loaded Dataset: {df.shape[0]} rows, {df.shape[1]} columns" )
```

```python
#print(df.shape)
```

```python
df.head()
```

```python
# SETP 1 : Header Cleaning(rename)
print(df.columns.tolist())
# all columns to lowercase and replace space with '_'
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')

print("After")
print(df.columns.tolist())
```

```python
#Step 2 : Type Conversion and Currency Cleaning
values_with_sign_dollar = df['spend'].astype(str).str.contains(r'\$')

print(df.loc[values_with_sign_dollar,['campaign_id','spend']].head(3))

#replace and character not a digit, dot or minus with nothing, only numbers allowed
df['spend'] = df['spend'].astype(str).str.replace(r'[^\d._]','',regex= True)
#convert to numeric
df['spend'] = pd.to_numeric(df['spend'])

print("After Fix")

print(df.loc[values_with_sign_dollar,['campaign_id','spend']].head(3))
```

```python
#STEP 3: CATEGORICAL TYPOS (FUZZY LOGIC)

#check for unique names in channel column
print(df['channel'].unique())

#replace with misspelled ones
cleanup_map = {
    'Tik_Tok' : 'TikTok',
    'Facebok' : 'Facebook',
    'Insta_gram' : 'Instagram',
    'Gogle' : 'Google Ads',
    'E-mail' : 'Email',
    'N/A ' : np.nan # handling ghost values
}

#channel column replace with cleanup_map
df['channel'] = df['channel'].replace(cleanup_map)

print("AfterFix")
print(df['channel'].unique())
```

```python
#Step4 : Handling mixed booleans
print(df['active'].unique())

bool_map = {
    'Y' : 'True',
    'Yes': 'True',
    '1':'True',
    1 : 'True',
    
    '0' : 'False',
    'No' : 'False',
    0 :'False'
    
}

df['active'] = df['active'].map(bool_map).fillna(False).astype(bool)

print("After Fixed")
print(df['active'].unique())
```

```python
#Step 5 : Date Parsing
print(df['start_date'].dtype)

df['start_date'] = pd.to_datetime(df['start_date'], errors = 'coerce')
df['end_date'] = pd.to_datetime(df['end_date'], dayfirst= True, errors = 'coerce')

print("After Fix")
print(df['start_date'].dtype)     
```

```python
df
```

```python
#remove duplicate columns
df = df.loc[:, ~df.columns.duplicated()]
df
```

```python
#step 6 : Logical Intergrity vs Impressions
impossible_mask = df['clicks'] > df['impressions']
print(df.loc[impossible_mask,['campaign_id','impressions','clicks']].head(3))

# check for any values in clicks high than Impressions
```

```python
#Step 7 : Logical Integrity (Time travel)
# to check any end date in the past compare to start date

time_travel_mask = df['end_date'] < df['start_date']
print(df.loc[time_travel_mask,['campaign_id','start_date','end_date']].head(3))
```

```python
df.loc[time_travel_mask,'end_date' ] = df.loc[time_travel_mask,'start_date' ] +pd.Timedelta(days = 30)

print("After Fix")

print(df.loc[time_travel_mask,['campaign_id','start_date','end_date']].head(3))
```

```python
#Step 8: Handling Outliers (Winsorizing)
Q1 = df['spend'].quantile(0.25)
Q3 = df['spend'].quantile(0.75)

IQR = Q3 - Q1
upper_limit = Q3 + (3 * IQR)
outlier_mask = df['spend'] > upper_limit
print(df.loc[outlier_mask,['campaign_id','spend']].head(3))

print("FIX APPLIED")
df.loc[outlier_mask, 'spend'] = upper_limit
print(df.loc[outlier_mask,['campaign_id','spend']].head(3))
```

```python
#Step 9 : String parsing

print(df['campaign_name'].head(3))

df['season'] = df['campaign_name'].str.extract(r'Q\d_([^_]+)_')

print("FIX APPLIED")

print(df[['campaign_name','season']].head(3))
```
